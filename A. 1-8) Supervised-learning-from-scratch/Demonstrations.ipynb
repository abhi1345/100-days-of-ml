## Data for Linear Regression

```python
dataset = {1:2, 2:4, 3:6, 7:14}
dataset2 = {5:3, 10:5, 15:7, 19:10}

lists = sorted(dataset.items()) # sorted by key, return a list of tuples
x, y = zip(*lists) # unpack a list of pairs into two tuples

plt.plot(x, y)
plt.title('Points from First Dataset')

lists = sorted(dataset2.items()) # sorted by key, return a list of tuples
x, y = zip(*lists) # unpack a list of pairs into two tuples

plt.plot(x, y)
plt.title('Linear Regression Datasets')
plt.xlim(0, 20)
plt.ylim(0, 20)
plt.legend(['First Dataset', 'Second Dataset'])
plt.show()
```

## Linear Regression Example

```python
model = LinearRegression()
function = model.gradientDescent(dataset, 0.01, 1000)
print("Dataset 1:")
print("Model equation: y = {}x + {}".format(function[0], function[1]))
print("Prediction for {} is {}\n".format(6, round(model.predict(6), 3)))
function = model.gradientDescent(dataset2, 0.001, 10000)
print("Dataset 2:")
print("Model equation: y = {}x + {}".format(function[0], function[1]))
print("Prediction for {} is {}".format(6, round(model.predict(6), 3)))
```

## kNN Example

```python
dataset = {
(2.327868056, 2.458016525) : -1,
(3.032830419, 3.170770366) : -1,
(4.485465382, 3.696728111) : -1,
(3.684815246, 3.846846973) : -1,
(2.283558563, 1.853215997) : -1,
(7.807521179, 3.290132136) : 1,
(6.132998136, 2.140563087) : 1,
(7.514829366, 2.107056961) : 1,
(5.502385039, 1.404002608) : 1,
(7.432932365, 4.236232628) : 1
}

ones = [k for k in dataset.keys() if dataset[k] == 1]
zeroes = [k for k in dataset.keys() if dataset[k] == -1]
plt.scatter(*zip(*ones))
plt.scatter(*zip(*zeroes))
plt.title('SVM Data')
plt.legend(['1', '-1'])
plt.show()
```

## SVM Example

```python
model = SupportVectorMachine()
model.train(dataset, 10)
tup = (6,2)
print("The prediction for 6,2 is {}.".format(model.predict(tup)))

testdata = {
(1, 4) : -1,
(3, 4) : -1,
(4, 3) : -1,
(5, 4) : -1,
(1, 1) : -1,
(9, 1) : 1,
(8, 4) : 1
}

#Accuracy Check
acc, div = 0, 0
for t in testdata.keys():
    if model.predict(t) == testdata[t]:
        acc += 1

    div += 1

print("This model has {}% accuracy on the test data.".format(100*acc/div))
```
