# Credit Card Fraud Detection with Boosted Decision Trees

```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load in 

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.ensemble import AdaBoostClassifier
# Input data files are available in the "../input/" directory.
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

dataset = pd.read_csv("/kaggle/input/creditcard.csv")
# Any results you write to the current directory are saved as output.
```

## Data

```python
dataset = dataset.sample(frac=1).reset_index(drop=True)
dataset.tail(3)
```

```python
X = dataset.drop(["Class"], axis=1)
y = np.array(dataset["Class"])

train_test_split = int(0.7*len(X))

train, test = dataset.iloc[:train_test_split], dataset.iloc[train_test_split:]

split_proportion = 0.7
train_test_split = int(split_proportion*len(X))
```

```python
trainX = train.drop(["Class"], axis=1)
trainY = train["Class"]
```

## Model Training

```python
model = AdaBoostClassifier(n_estimators=500)
model.fit(trainX, trainY)
```

## Error Analysis

```python
# True Positive Rate
testOnes = test[test["Class"] == 1]
testOnesX = testOnes.drop("Class", axis=1)
testOnesY = testOnes["Class"]

model.score(testOnesX, testOnesY)
```

```python
# True Negative Rate
testZeros = test[test["Class"] == 0]
testZerosX = testZeros.drop("Class", axis=1)
testZerosY = testZeros["Class"]

model.score(testZerosX, testZerosY)
```

```python
# Overall Accuracy
testX, testY = test.drop(["Class"], axis=1), test["Class"]
model.score(testX, testY)
```
